```python
import pandas as pd
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# Alterando o conteudo da linha
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False)
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False)
```

```python
vendas.info()
```

```python
#Tratando casos nulos
vendas["estado"].fillna("MS", inplace=True)

# Verificar informações do DataFrame após a alteração
vendas.info()
```

```python
media = vendas["Preço"].mean()
```

```python
vendas["Preço"] = vendas["Preço"].fillna(media)
vendas.info()
```

```python
vendas_correta = vendas.query("Preço_com_frete > Preço")
vendas_correta.info()
```

```python
vendas_correta.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False)
```

```python
  #Arredondar
  round(vendas_correta.groupby(["Nome_Departamento"])["Preço_com_frete"].agg("mean").sort_values(ascending = False),2)
```

```python
vendas_correta["mes"] = vendas_correta["Data"].dt.month
```

```python
vendas_correta.head()
```

```python
vendas_correta.groupby("mes").idcompra.nunique().sort_values(ascending =False)
```

```python
cliente = pd.read_excel("/content/cliente_varejo.xlsx")
```

```python
cliente.head()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda": float})
```

```python
cliente.info()
```

```python
#Unindo tabelas
```

```python
vendas_correta = vendas_correta.merge(cliente, how= "left", on="cliente_Log")
```

```python
vendas_correta.head()
```

```python
agg_canal_venda = round(vendas_correta.groupby(["idcanalvenda"])["renda"].agg("mean").reset_index().sort_values(by="renda",ascending = False),2)
agg_canal_venda
```

```python
agg_ban_idade = round(
    vendas_correta.groupby("bandeira")["idade"].mean().reset_index().sort_values(by="idade", ascending=False),
    2
)
agg_ban_idade
```

```python
  # Usando a matplotlib.pyplot
  import matplotlib.pyplot as plt
```

```python
agg_ban_idade.info()
```

```python
plt.figure(figsize=(10,8))
plt.bar(agg_ban_idade["bandeira"],agg_ban_idade["idade"],color="blue")
plt.ylabel ("Média de idade")
plt.xlabel("Bandeira")
plt.title("Idade por bandeira", loc="center",fontsize = 18,fontweight = "bold",color = "black")

# Exibindo o gráfico
plt.show()
```

```python
agg_canal_venda
```

```python
plt.figure(figsize=(10,8))
plt.bar(agg_canal_venda["idcanalvenda"],agg_canal_venda["renda"],color="green")
plt.ylabel ("Média de Renda")
plt.xlabel("Canal de Venda")
plt.title("Renda por Canal", loc="center",fontsize = 18,fontweight = "bold",color = "black")

# Exibindo o gráfico
plt.show()
```

```python
agg_data = vendas_correta.groupby(["Data"]).idcompra.nunique().reset_index()
agg_data
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("Data")
plt.ylabel("Qtd de vendas")
x = agg_data["Data"]
y = agg_data["idcompra"]
plt.plot(x,y,color = "pink" , linewidth=2)
# Exibindo o gráfico
plt.show()
```

```python
agg_dept_pre = round(vendas_correta.groupby(["Nome_Departamento"])["Preço_com_frete"].agg("mean").reset_index().sort_values(by = "Preço_com_frete",ascending = False),2)
agg_dept_pre
```

```python
plt.figure(figsize=(10,8))
plt.xlabel("Departamento")
plt.ylabel("Preco")
x = agg_dept_pre["Nome_Departamento"]
y = agg_dept_pre["Preço_com_frete"]
plt.plot(x,y,color = "pink" , linewidth=2)
# Exibindo o gráfico
plt.show()
```

```python
plt.figure(figsize=(30,25))
plt.bar(agg_dept_pre["Nome_Departamento"] ,agg_dept_pre["Preço_com_frete"] ,color="purple")
plt.ylabel ("Preco")
plt.xlabel("Departamento")
plt.title("Preço por Departamento", loc="center",fontsize = 18,fontweight = "bold",color = "black")

# Exibindo o gráfico
plt.show()
```
